```python
!pip install --quiet openai

import os
from openai import OpenAI
os.environ["OPENAI_API_KEY"] = "#####"

client = OpenAI()
MODEL = "gpt-4o-mini"
```

```python
def summarize_text(text: str) -> str:
    """Initial summary."""
    resp = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": "You are an expert summarizer."},
            {"role": "user", "content": f"Summarize this text in about 100 words:\n\n{text}"}
        ]
    )
    return resp.choices[0].message.content.strip()

def critique_summary(original_text: str, summary: str) -> str:
    """Ask AI to critique its own summary."""
    resp = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": "You are a critical writing coach."},
            {"role": "user", "content": f"Original text:\n{original_text}\n\nSummary:\n{summary}\n\nCritique the summary for accuracy, clarity, and completeness. Point out missing or misleading points."}
        ]
    )
    return resp.choices[0].message.content.strip()

def improve_summary(original_text: str, critique: str, old_summary: str) -> str:
    """Generate an improved version based on critique."""
    resp = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": "You are a skilled editor improving AI writing."},
            {"role": "user", "content": f"Original text:\n{original_text}\n\nOld summary:\n{old_summary}\n\nCritique:\n{critique}\n\nWrite a new, improved summary that fixes the issues noted."}
        ]
    )
    return resp.choices[0].message.content.strip()
```
