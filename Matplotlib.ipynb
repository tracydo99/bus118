```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

see distribution and frequency. for example, how many students got As Bs or Cs, or scores on a test

```python
x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.1
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

show relationship between 2 variables. for example, scores of 2 students, one who studied 2 hours and one who studied 10 hours.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.show()
```

outliers, median, quartiles. useful for visualizing grades/scores

```python
time = np.arange(0, 10, 0.1)
amplitude = np.sin(time)
plt.plot(time, amplitude)
plt.title("Line Plot")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

shows trends over time for sequential data. for example, stock market trends

• plt.savefig() must be called before plt.show() — once you show the plot, the figure is
closed.
• You can specify the resolution with dpi= (dots per inch), especially useful for print-quality
images.
• Supported formats include: .png, .jpg, .jpeg, .pdf, .svg, .eps, .tif.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Sine Wave Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True)

# Save in different formats
plt.savefig('sine_plot.pdf')     # Save as PDF (vector format)
plt.savefig('sine_plot.png')     # Save as PNG
plt.savefig('sine_plot.jpg', dpi=300)  # Save as JPEG with high resolution

# Show the plot (optional)
plt.show()
```
